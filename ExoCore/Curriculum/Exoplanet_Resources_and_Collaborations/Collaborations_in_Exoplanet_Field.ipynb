_**1. TESS Object of Interest (TOI)**_

**Overview**

The Transiting Exoplanet Survey Satellite (TESS) aims to discover thousands of exoplanets orbiting the nearest, brightest stars in the sky. The primary types of exoplanets the TESS project is searching for are Earth-like planets and super-Earths. The TESS Object of Interest (TOI) catalog contains discoveries identified by the TESS project and community as objects of interest. This catalog is updated approximately once a week according to the ExoFOP-TESS TOI list. As of February 2024, over 7,000 objects of interest have been logged. This list can be found in the [exoplanet archive](https://exoplanetarchive.ipac.caltech.edu/cgi-bin/TblView/nph-tblView?app=ExoTbls&config=TOI).

_**2. TESS Follow-Up Observing Program (TFOP)**_

Another main goal of TESS is to measure the masses of 50 planets with radii smaller than 4 Earth radii. TFOP is a group of observers tasked with following up on these measurements and TOIs. Stars hosting high-priority planet candidates will be reviewed to confirm the presence of planets, refine their radii, and measure their masses.

TFOP is overseen by the Smithsonian Astrophysical Observatory (SAO) and MIT, in accordance with the TESS Space Office (TSO). The goals of TFOP are to facilitate communication and efficacy within the TESS Science Team and the exoplanet community. This is intended to strengthen scientific knowledge and prevent the duplication of effort.

TFOP consists of five branches: Seeing Limited Photometry, Recon Spectroscopy, High-Resolution Imaging, Precise Radial Velocities, and Space Photometry. Observations will be performed with committed time on the Las Cumbres Observatory Global Telescope Network (for photometry and spectroscopy), the MEarth (photometry) and TRES (spectroscopy) observatories, as well as several other provisions. 

Within TFOP is a web-based service, ExoFOP, generated by the NASA Exoplanet Science Institute (NExScI). ExoFOP is essential in mediating TFOP ventures, observations, and resources. ExoFOP is a medium through which the community can share data and information on follow-up observations to aid in the efficiency of community telescope resources. 

ExoFOP offers a dedicated page for TESS, known as ExoFOP-TESS. ExoFOP-TESS currently includes pages for each of the 10.8 million stars in the TESS Candidate Target list (CTL). These pages hold data from the TESS Input Catalog (TIC) Version 5, including coordinates, magnitudes, stellar parameters, and planet parameters (if available).

ExoFOP-TESS will offer:

- Summaries of TESS-observed targets
- Overview of individual targets
- Summary of observations
- Observing notes
- Links to outside services (Exoplanet Archive tools, MAST data and documentation, Simbad and other host star resources, community-produced light curves and products, etc.)
- Telescope resource calendar


_**3. James Webb Space Telescope Early Release Program (JWST ERS)**_

Collecting data with space observatories can be challenging. Still, the James Webb Space Telescope (JWST) provides high-precision time-series observations that have steepened the learning curve for some users. As observation slots were awarded to researchers in 2018 for JWST Cycle 1, it was (and still is) important that a team of collaborators assist in this transition. 

The James Webb Space Telescope Early Release Science program (JWST ERS) aims to provide the astronomical community with readily available datasets and analysis toolkits before Cycle 2 observations, as well as to assist in collaborations focused on learning how to make the most of JWST for studying exoplanets.

**There are three primary objectives of the program:**

* Determine the spectrophotometric timeseries performance of the key instrument modes on timescales relevant to transits for a representative range of target star brightnesses. We will perform a quantitative assessment of the data quality to inform future instrument and integration/exposure time selections and shed light on the feasibility of characterizing potentially habitable exoplanets.

* Jump-start the process of developing remediation strategies for instrument-specific systematic noise. We will formulate open-source data analysis code that provides a foundation for establishing best practices in removing systematics.

* Provide the community with a comprehensive suite of transiting exoplanet data to fully demonstrate JWST's scientific capabilities in this area. We will leverage these datasets to engage the community in an open data challenge that will deepen our understanding and accelerate the diffusion of this knowledge.

The program is to conduct three scientific and technical investigations that, together, create six separate observations. These three investigations are known as the Panchromatic Transmission, MIRI Phase Curve, and Bright Star Programs, and the observations utilize five of JWST's spectroscopic modes. The Panchromatic Transmission Program involves four transit observations of a single target with a total of three instruments: NIRISS (one observation with the SOSS mode), NIRSpec (one observation each in the G235H and G395H modes), and NIRCam (one observation in the F322W2 mode). The MIRI Phase Curve Program includes a full-orbit phase curve of a planet obtained with MIRI in the LRS slitless mode. A community process was established to improve the program's efficiency. The community process involved open email, message board, and telecon discussions, and democratic voting was used to make decisions. The resulting program is comprehensive, leveraging the three transiting-planet geometries (primary transit, secondary eclipse, and phase curves), covering all atmospheric observables (compositions, thermal structures, and dynamics on both the day- and night-sides), and utilizes all four JWST instruments.

The program is efficient because it includes only the highest-priority observations identified by the community process. Other possible observations were considered of lower priority due to the ERS goals and were released from this program. For example, repeated observations of the same target with the same setup on the near-infrared (NIR) instruments were considered a lower priority because we are already observing the same target with all three of the NIR instruments and four unique settings. By requiring a consistent transit shape (modulo limb darkening) for each observation, cross-comparison of these data sets will result in the same test as repeated observations with the same setup.

The management structure of the Transiting Exoplanet Community ERS program is based on nominal practices dating back to the Kepler mission. It is designed to enable a large, distributed, and diverse team to deliver products on time. The PI (person(s) in charge) and co-PIs were peer-elected. The Science Council advises the project leadership and provides mediation for conflict resolution. Altogether, the team offers diversity of expertise (58% observers and 33% theorists, with 9% self-identified as "other", which is a category that includes instrumentalists and administrators) and geography (54% US, 46% EU & Canada), with members from research centers, research universities, and smaller undergraduate-focused institutions. Gender balance is 23% women across the team (near the national average for astronomy), and 44% women at leadership levels.

The JWST ERS program is essential in accelerating a comprehensive census of transiting exoplanets. By an open data structure, there is a dramatic advance in our understanding of planetary nature, origins, climate, atmospheric physics and chemistry, and habitability. The launch and continued use of JWST will elevate exoplanet research to levels we've never achieved. With this program, we are on track to achieve such ambitions.

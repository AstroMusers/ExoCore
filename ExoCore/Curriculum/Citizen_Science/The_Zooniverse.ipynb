## Overview <a name="Overview"></a>



The Zooniverse is a prime example of how citizen science can leverage the power of public participation from nonprofessionals. Today, projects on The Zooniverse are not limited to the field of exoplanet or even astrophysics research. The platform is suitable for projects in any field of research and hosts a diverse array of citizen science projects. A research project in any field requires some sort of classification or information processing technique. This information processing can be simple but heavily repetitive and mundane when dealing with extensive data sets. For example, information processing could involve the characterization of a galaxy based on its shape in an image or identifying an animal from a video. Volunteers can help researchers with this part of the research process by participating in a Zooniverse project to characterize information and gather data. The projects on The Zooniverse typically have large volumes of data that can be time-consuming for a research team to sort through and too difficult for computers to analyze accurately. When many volunteers join together using their naturally human pattern recognition skills, the efficiency of a research project improves. The Zooniverse has led to scientific discoveries and advancements through public engagement in science. 

In terms of exoplanet research, The Zooniverse hosts many related projects. The project-building feature can be handy for exoplanet researchers. For example, you could use volunteers to identify transits. Later lessons in this module will discuss a few projects that have used citizen science for this purpose.

## Building A Project on Zooniverse <a name="Project_Builder"></a>
The most helpful feature of Zooniverse for exoplanet research is its project builder. The Zooniverse website provides an accessible and comprehensive tutorial on its project-building feature. With the Project Builder tool, researchers can create a new project by providing information about the scientific objectives, research questions, and datasets they wish to analyze. They can open up a dataset and specify the tasks volunteers will perform, such as classifying images, transcribing text, or identifying patterns in data.

Select "Create a new project" to get started on your project.


### Project Details
Zooniverse prompts the user to input details on their project, including a name, description, introduction, workflow description, researcher quote, announcement banner, discipline tag, other tags, and external links. Properly outlining the project through these prompts allows researchers to attract and direct volunteers in a manner specific to the individual project. To appeal to volunteers, it is essential to develop the project details section thoroughly. When working on the project details, let volunteers know why the project is vital and exciting. Make the project attractive by using the feature to drop in an avatar and background image. The project details section will be a volunteer's first impression of your project. 

### About
Following the project details section, the researcher is provided five open fields to describe the project's goals to volunteers in the About section. This section fully outlines the research, team, results, and education behind the project, along with the opportunity to include an FAQ page. These fields render markdown so that you can easily add media and hyperlinks. Appeal to volunteers with graphics and descriptions of the people on your team. You may leave any of these fields blank, but describe all aspects of your current project fully. Avoid jargon, so the description is comprehensive to citizen volunteers.

### Collaborators
Researchers can add collaborators to their projects and classify their roles based on the extent to which they will contribute to the project. This feature enables collaboration through the Zooniverse platform, so you can split the project development across different team members.

### Field Guide
The field guide tab allows researchers to upload a field guide that helps volunteers complete the project classifications. 

### Tutorial
Researchers should upload a short tutorial so that volunteers can quickly and easily start contributing to the project. 

### Media
Media can be uploaded to support the project.

### Visibility
While building the project, researchers can change its visibility status. It can be public, so anyone with the URL can access it, or it can be private and visible only to its collaborators. The project status can also be changed between development and live under the visibility tab.

### Talk
Researchers and collaborators can attach discussion boards to the project under the Talk tab, allowing collaborators and volunteers to communicate with each other. While it is possible to attach many boards to a project, it is recommended to first set up the default board. After that, other boards can be established for topics like "Questions" or "Technical Support." The Talk section of the project builder establishes how you allow communications between volunteers and researchers. 

### Data Exports
Project data can be exported in CSV format through the Data Exports tab, and discussions can be exported in JSON format. Sometimes, project data can take a significant amount of time to export. Additionally, only one type of export can be requested within a 24-hour period. 

### Workflows
Researchers develop the workflows containing the tasks they wish for volunteers to complete. A new workflow can be added by clicking the new workflow button under the Workflows tab. Inside a created workflow, one can create tasks in a question, drawing, text, or survey format. This allows for easy customization and truly makes the Zooniverse suitable for many different types of projects. Projects can have multiple workflows, and each should be associated with a tutorial developed under the Tutorials tab. Volunteers can see active workflows on the project's statistics page.

### Subject Sets
Researchers upload subject sets of data for the volunteers to classify. Many different file types for manifests and subject images are supported. 


Overall, the Project Builder tool on Zooniverse empowers researchers to create and manage their citizen science projects, enabling them to leverage the collective efforts of volunteers to address complex scientific questions, analyze large data sets, and advance research in various fields of study. [Here](https://mast.stsci.edu/phad/about.html) is another tutorial The Zooniverse provides on their website.

## Communications and News <a name="Communications"></a>
Under [Zooniverse Talk](https://www.zooniverse.org/talk), one can find many discussion boards related to individual projects and the site itself. These discussion boards allow for communication between professional researchers and volunteers. Additionally, such open discussion boards allow citizen science volunteers to answer each other's questions and provide help without the need for professional researchers to always get involved. Discussion boards are a great place to post announcements about a project, gather interest from potential volunteers, and demonstrate project progress. 


The Zooniverse also keeps its community up-to-date with project and site news in the [Daily Zooniverse](https://daily.zooniverse.org/) and their [blog](https://blog.zooniverse.org/?_ga=2.63500527.1771494547.1716257124-1937268864.1716257124). 

## Publications and Project Outcomes <a name="Project_Outcomes"></a>
Citizen science projects on the Zooniverse result in research papers and open-source sets of analyzed data. Many vital discoveries occur due to volunteer classification and discussions between citizen scientists and researchers. For example, a paper was published in The Astronomical Journal in April 2024 after volunteers for a Zooniverse project called Planet Hunters TESS resulted in the discovery of a Mini-Neptune in the habitable zone orbiting around a binary star system (Eisner et al. 2024). Citizen scientists classified two of the three identified transit events that resulted in the discovery. The Zooniverse keeps track of publications related to discoveries from Zooniverse projects [here](https://www.zooniverse.org/about/publications).

## Overview <a name="Overview"></a>

Citizen science projects can be funded through various methods, including federal grants, crowdfunding, and public fundraising. Without adequate funding and resources, the project's scope and impact will likely be reduced. It may be difficult to attract sufficient volunteer participation if the project development is limited by funding. 

## NSPIRES <a name="NSPIRES"></a>



Developed citizen science projects can apply for federal funding via NASA's Research Opportunities in Space and Earth Science (ROSES) via the Solicitation and Proposal Integrated Review and Evaluation System (NSPIRES). The predominant funding entity for citizen science projects is the [F.9 Citizen Science Seed Funding Program](https://nspires.nasaprs.com/external/solicitations/summary!init.do?solId=%7BA08B277F-1BFE-4663-3E6F-C178EBA87C8C%7D&path=open). The F.9 Citizen science Seed Funding Program is designed to support the creation of citizen science methods in research. The program nurtures projects during their initial launch or transition phases.

## Overview <a name="Overview"></a>
Planet Hunters TESS (PHT) is a prominent citizen science project that invites volunteers from around the world to participate in the search for exoplanets. PHT is part of the Zooniverse which hosts a variety of different citizen science programs. The platform takes data from the Transiting Exoplanet Survey Satellite (TESS) which is now an extended mission. During its prime mission, TESS surveyed about 75% (NASA, 2021). PHT has used volunteers to identify exoplanet candidates from these surveys. Volunteers analyze the light curves of surveyed stars looking for dips that could indicate a transiting exoplanet.

Participating in exoplanet classification on PHT can provide new researchers with practical, hands-on experience in analyzing real astronomical data. By working with actual data from space telescopes like TESS, new researchers can become familiar with the types of observations and measurements commonly used in exoplanet studies. Engaging in citizen science projects like PHT provides new researchers with opportunities to connect with experienced scientists and fellow enthusiasts in the field. Planet Hunters TESS is a prominent citizen science project in the field of exoplanet research.

## [The Planet Hunters Analysis Database (PHAD)](https://mast.stsci.edu/phad/about.html) <a name="PHAD"></a>
PHAD is the public database listing the classifications made by volunteers working on the Planet Hunters TESS Project. The data is accessible through the linked website or programmatically through Jupyter Notebooks on the Zooniverse Github page. All of this data can be used and analyzed by researchers like you to answer your research questions.

#### Accessing Data on the Website
On the website, there are two different tabular views. You can learn more about these two different views and the PHAD website [here](https://mast.stsci.edu/phad/about.html).

Transit view organizes subjects based on their subject IDs.


User view organizes subjects based on the IDs of the users who identifid them.


Both viewing options convey most the same information with three informational columns: 
- Center
- Width
- Weighted Columns

The two different tabular views only differ in the sense that the transit view contains information on user count (how many users have identified the transit), while the user oriented view only displays data that has been identified by a sufficient number of users. Not all data on fromn the Planet Hunters calssification website is listed in either table. The tables only contain subjects that have been identified as transits by enough reliable viewers. This data can all be used in your own projects as long as you credit the Zooniverse volunteers and Planet Hunters team.

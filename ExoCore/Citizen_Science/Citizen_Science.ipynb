1. Introduction


Citizen science initiatives have significantly contributed to the advancing field of exoplanet research by harnessing the collective power of public participation to analyze datasets. Exoplanet research often entails sorting through extensive data collected by space telescopes like Kepler, TESS, and others. Notable projects such as Galaxy Zoo, Planet Hunters, and Backyard Worlds on the Zooniverse platform allow volunteer participants to identify and characterize possible exoplanet transits or anomalies in light curves. This volunteer involvement complements automated algorithms, enhancing the accuracy of potential exoplanet candidates that might otherwise be overlooked by computers alone. Citizen science provides opportunities for novices to familiarize themselves with exoplanet research.


There are three main ways in which citizen science platforms can benefit new professionals entering the field of exoplanet researchers.
- Identifying and characterizing data for exposure to exoplanet data processing
- Opening up a data set and allowing the public to characterize it
- Utilizing data for independent projects


The following sections will contain discussion on some of the main citizen science platforms in the field currently.






2. The Zooniverse
Zooniverse is a platform that hosts various citizen science projects. Zooniverse projects typically involve large datasets that are difficult for computers to analyze accurately, but humans can easily recognize patterns or anomalies. Volunteers on Zooniverse help process these datasets by performing tasks such as categorizing images, marking objects of interest, or transcribing text. The projects on Zooniverse cover a wide range of disciplines, including astronomy, biology, ecology, climate science, and more. Zooniverse has the ability to leverage the collective power of a large number of volunteers to accomplish tasks that would be impractical or impossible for a small team of researchers to do on their own. This model has led to scientific discoveries and advancements through providing opportunities for public engagement in science.

A. Galaxy Zoo
The Zooniverse kicked off with its first project, Galaxy Zoo, in 2007 with a data set of a few million galaxies from the Sloan Digital Sky Survey. While the Sloan Digital Sky Survey still provides many of the images on the site today, Galaxy Zoo has added images from more distant sky surveys including the Hubble Space Telescope. New data is now being added to Galaxy Zoo rapidly. Recent inclusions incorporate images from the United Kingdom Infrared Telescope (UKIRT), the Dark Energy Camera Legacy Survey (DECaLS), and the Galaxy and Mass Assembly Survey (GAMA). Galaxy Zoo recently classified a dataset of fully simulated galaxies from the Illustris project for the first time. Along with the additions to Galaxy Zoos original form, the platform has omitted completed images that are no longer useful to researchers. Past iterations of the platform can be accessed through online archives.


Galaxy Zoos diverse array includes surveys on our local group and distant galaxies. By surveying galaxies at different distances, Galaxy Zoo enables researchers to study how galaxies have evolved over time. Comparing local galaxies to those farther away provides researchers with insight into the processes driving galaxy formation and evolution throughout the ages of the universe. As data from more distant galaxies is added to the platform, questions about the relationship between galaxy distance and classification have arisen. In order to establish any dependence of classification on distance, Galaxy Zoo has simulated images of galaxies from different viewing perspectives. Analysis of these images will help discern the effects of redshift on galaxy classification.
  


B.  Planet Hunters TESS
Planet Hunters TESS (PHT) is a prominent citizen science project that invites volunteers from around the world to participate in the search for exoplanets. PHT is part of the Zooniverse which hosts a variety of different citizen science programs. The platform takes data from the Transiting Exoplanet Survey Satellite (TESS) which is now an extended mission. During its prime mission, TESS surveyed about 75% (NASA, 2021). PHT has used volunteers to identify exoplanet candidates from these surveys. Volunteers analyze the light curves of surveyed stars looking for dips that could indicate a transiting exoplanet.


Participating in exoplanet classification on PHT can provide new researchers with practical, hands-on experience in analyzing real astronomical data. By working with actual data from space telescopes like TESS, new researchers can become familiar with the types of observations and measurements commonly used in exoplanet studies. Engaging in citizen science projects like PHT provides new researchers with opportunities to connect with experienced scientists and fellow enthusiasts in the field.


PHT makes it extremely easy for volunteers to get involved in classification.
- Visit the program site
- Click classify
- Complete the 10 minute tutorial on analyzing light curves
- Discover exoplanets

C. Backyard Worlds
A final notable project on Zooniverse is Backyard Worlds: Planet 9, a project aimed at discovering new celestial objects in our solar system. Backyard Worlds engages citizen scientists in the search for new objects by having them analyze images taken by NASA's Wide-field Infrared Survey Explorer (WISE) telescope. Participants examine these images to identify moving objects, such as brown dwarfs, low-mass stars, and potentially even the elusive Planet Nine. These celestial bodies can be difficult to detect due to their faintness and distant location, but with the help of volunteers, the project aims to uncover them.


D. Building a project in Zooniverse
While participating in Zooniverse projects like Galaxy Zoo, Planet Hunters, and Backyard Worlds can be beneficial to those entering the field of exoplanet research, perhaps the most substantial feature of Zooniverse, is its project builder. Zooniverse provides an accessible and comprehensive tutorial on its project building feature. With the Project Builder tool, researchers can create a new project by providing information about the scientific objectives, research questions, and datasets they wish to analyze. They can open up a dataset and specify the type of tasks volunteers will perform, such as classifying images, transcribing text, or identifying patterns in data.


Project Details
Zooniverse prompts the user to input details on their project including a name, description, introduction, workflow description, researcher quote, announcement banner, discipline tag, other tags, and external links. Properly outlining the project through these prompts allows researchers to attract and direct volunteers in a manner specific to the individual project


About
Following the project details section in the project builder, the researcher has an open field to describe the goals of the project to volunteers. Researchers are encouraged to avoid jargon, so the description is comprehensive to citizen volunteers.

Introduction

Citizen science initiatives have significantly contributed to the advancing field of exoplanet research by harnessing the collective power of public participation to analyze datasets. Exoplanet research often entails sorting through extensive data collected by space telescopes like Kepler, TESS, and others. Notable projects such as Planet Hunters and Backyard Worlds allow volunteer participants to identify and characterize possible exoplanet transits or anomalies in light curves. This volunteer involvement complements automated algorithms, enhancing the accuracy of potential exoplanet candidates that might otherwise be overlooked by computers alone. Citizen science provides opportunities for novices to familiarize themselves with exoplanet research.

There are three main ways in which citizen science platforms can benefit new professionals entering the field of exoplanet researchers.
Identifying and characterizing data for exposure to exoplanet data processing
Opening up a data set and allowing the public to characterize it
Utilizing data for independent projects

The following sections will contain discussion on some of the main citizen science platforms in the field currently.

1. Planet Hunters TESS

Planet Hunters TESS (PHT) is a prominent citizen science project that invites volunteers from around the world to participate in the search for exoplanets. PHT is part of the Zooniverse which hosts a variety of different citizen science programs. The platform takes data from the Transiting Exoplanet Survey Satellite (TESS) which is now an extended mission. During its prime mission, TESS surveyed about 75% (NASA, 2021). PHT has used volunteers to identify exoplanet candidates from these surveys. Volunteers analyze the light curves of surveyed stars looking for dips that could indicate a transiting exoplanet.

Participating in exoplanet classification on PHT can provide new researchers with practical, hands-on experience in analyzing real astronomical data. By working with actual data from space telescopes like TESS, new researchers can become familiar with the types of observations and measurements commonly used in exoplanet studies. Engaging in citizen science projects like PHT provides new researchers with opportunities to connect with experienced scientists and fellow enthusiasts in the field.

While participation in transit identification can be beneficial to those entering the field of exoplanet research, perhaps the most substantial feature of Zooniverse, PHT’s host program, is its project builder. PHT provides an accessible and comprehensive tutorial on its project building feature. With the Project Builder tool, researchers can create a new project by providing information about the scientific objectives, research questions, and datasets they wish to analyze. They can open up a dataset and specify the type of tasks volunteers will perform, such as classifying images, transcribing text, or identifying patterns in data. 
Building a project in Zooniverse

Project Details

Zooniverse prompts the user to input details on their project including a name, description, introduction, workflow description, researcher quote, announcement banner, discipline tag, other tags, and external links. Properly outlining the project through these prompts allows researchers to attract and direct volunteers in a manner specific to the individual project

About

Following the project details section in the project builder, the researcher has an open field to describe the goals of the project to volunteers. Researchers are encouraged to avoid jargon, so the description is comprehensive to citizen volunteers.




Collaborators

Researchers can add collaborators to their projects and classify their roles based on the extent that they will contribute to the project. 


Field Guide

The field guide tab allows researchers to upload a field guide.

Tutorial

Researchers should upload a short tutorial, so that volunteers can quickly and easily get started on contributing to the project.

Media

Media can be uploaded to support the project.

Visibility

While building the project, researchers are able to change the visibility status. The project can be public, so anyone with the URL can access it, or it can be private and visible only to its collaborators.




Talk

Researchers and collaborators can attach discussions to the project, allowing collaborators and volunteers to communicate with each other. 

Data exports

Project data can be exported through the Data Exports tab in CSV format. Discussions can be exported in JSON format.


Workflows

Researchers instruct volunteers on the tasks they want volunteers to perform through workflows. 

Subject Sets

Researchers can create subject sets of data for the volunteers to analyze. 

Overall, the Project Builder tool on Zooniverse empowers researchers to create and manage their own citizen science projects, enabling them to leverage the collective efforts of volunteers to address complex scientific questions, analyze large data sets, and advance research in various fields of study.

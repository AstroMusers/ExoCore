```python
import matplotlib.pyplot as plt

# Sample data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 6]

# Plotting the graph
plt.plot(x_values, y_values, marker='o', linestyle='-')

# Adding title and labels
plt.title('Sample Line Graph')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the graph
plt.grid(True)
plt.show()
```
